```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
from pandas.plotting import register_matplotlib_converters
from sklearn.model_selection import train_test_split
import urllib
import os
import csv
import cv2
import time
from PIL import Image
import shutil
from time import gmtime, strftime

RANDOM_SEED = 42

np.random.seed(RANDOM_SEED)
```

# Training



```python
from imageai.Detection.Custom import DetectionModelTrainer

%cd ../../../
trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/WiderPerson/ImageAI/output")
trainer.setTrainConfig(object_names_array=["pedestrian"], batch_size=8,
                       num_experiments=8, train_from_pretrained_model="pretrained-yolov3.h5")
trainer.trainModel()
```

```python
trainer.evaluateModel('/content/WiderPerson/ImageAI/output/models/detection_model-ex-001--loss-0052.153.h5', 
                      '/content/WiderPerson/ImageAI/output/json/detection_config.json', 
                      object_threshold=0.5)
```

```python
from imageai.Detection.Custom import CustomVideoObjectDetection

execution_path = os.getcwd()

video_detector = CustomVideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath("/content/WiderPerson/ImageAI/output/models/detection_model-ex-001--loss-0053.397.h5")
video_detector.setJsonPath("/content/WiderPerson/ImageAI/output/json/detection_config.json")
video_detector.loadModel()

video_detector.detectObjectsFromVideo(input_file_path="/content/WiderPerson/2016_04_08_13_49_05FrontColor.avi",
                                          output_file_path=os.path.join(execution_path, "bus_detected"),
                                          frames_per_second=20,
                                          minimum_percentage_probability=30,
                                          log_progress=True)
```
