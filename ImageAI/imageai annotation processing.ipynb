```
import os 
import pandas as pd
from PIL import Image
from pascal_voc_writer import Writer
```

```
colnames = ['image_name', 'x1', 'y1', 'x2', 'y2', 'label']
```

```
def generate_xml_annotations(annotations_csv_path, image_path):
    ''' Generate xml files in Pascal VOC format to train with imageAI Yolov3
    Arguments: 
        --annotations_csv_path: Path to the csv file which was created to train 
                                in csv format (e.g. keras-retinanet)
        --image_path: Path to the images where stored right now, not during
                      training time
    
    '''

    df = pd.read_csv(annotations_csv_path, header= None, names=colnames)
    df['image_name'] = df.image_name.str.slice(-10)

    for image_name in df['image_name'].unique():
        im = Image.open(image_path + image_name)
        w, h = im.size
        writer = Writer('WiderPerson/ImageAI/output/' + image_name, w, h)

        for row in df.loc[df.image_name == image_name].iterrows():
            writer.addObject(row[1].label, row[1].x1, row[1].y1,
                                           row[1].x2, row[1].y2)

        writer.save(image_path + image_name[0:-4] + '.xml')
```
